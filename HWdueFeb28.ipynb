7. Make a bar graph of the number of words starting with each letter. You will have to import at least one more package to do this. Label the axes and give the whole plot a title.

8. What are the six longest words in the word list?

```python
word_length = [len(word) for word in wordlist]

word_data= list(zip(wordlist, word_length))

order_of_length = sorted(word_data, key=lambda x:x[1], reverse= True)

print ("The six longest words in wordlist are:" , order_of_length[:6])
print (       )

word_length2 = [len(word) for word in alpha_list]

word_data= list(zip(alpha_list, word_length2))

order_of_length = sorted(word_data, key=lambda x:x[1], reverse= True)

print ("The six longest words in alpha_list are:" , order_of_length[:6])
```
